### Adding dependencies for the extract, transform and load processes.

```python
#Dependencies
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

### The resources includes three CSVs files namely, Asylum_seekers, demographics and time_series. These are loaded into the dataframes using pandas.

```python
#Extracting data from csv files
asylum_seekers_monthly =pd.read_csv("Resources/asylum_seekers_monthly.csv")
asylum_seekers_monthly.sample(20)
asylum_seekers_monthly = asylum_seekers_monthly.rename(columns ={"Country / territory of asylum/residence":"host_country", 
                                                                "Origin":"origin", "Year":"year", "Month":"month", "Value":"value"})
asylum_seekers_monthly
```

### Data clean up to remove the special characters and null values in all the three datasets

```python
#Replacing the special characters with zero.
asylum_seekers = asylum_seekers_monthly.copy()
asylum_seekers=asylum_seekers.replace(['*'], [0])
asylum_seekers.sample(20)
```

```python
#Retrieving the column names for reference
asylum_seekers.columns
```

```python
#Reading the second dataset using pandas
demographics = pd.read_csv("Resources/demographics.csv")
demographics.sample(20)
```

```python
#Replacing the special characters and null values with zero.
demographics = demographics.fillna(0)
demographics=demographics.replace(['*'], [0])
demographics=demographics.replace([''], [0])
demographics.sample(20)
```

```python
#Renaming the column names in the standard format that is required in the database
labels= ['year', 'host_country', 'location_name',
       'female_0_4', 'female_5_11', 'female_5_17', 'female_12_17',
       'female_18_59', 'female_60plus', 'f_unknown', 'f_total', 'male_0_4',
       'male_5_11', 'male_5_17', 'male_12_17', 'male_18_59', 'male_60plus',
       'm_unknown', 'm_total']
demographics.columns = labels
```

```python
#Retrieving the column names for reference
demographics.columns
```

```python
#Dropping unwanted columns
demographics.drop(['female_0_4','female_5_11','female_5_17','female_12_17',
       'female_18_59', 'female_60plus', 'f_unknown','male_0_4',
       'male_5_11', 'male_5_17', 'male_12_17', 'male_18_59', 'male_60plus',
       'm_unknown', 'location_name'],axis =1,inplace = True)
demographics
```

```python
#Verifying the datatypes of the columns 
demographics.dtypes
```

```python
#Converting the total columns to integer to get the sum in the groupby function
demographics['f_total'] =pd.to_numeric(demographics['f_total'])
demographics['m_total'] =pd.to_numeric(demographics['m_total'])
demographics.dtypes
```

```python
#Grouping by year and host country to arrive at the total male and female asylum seekers
group_year_demographics = demographics.groupby(['year','host_country']).sum().reset_index()
print(group_year_demographics)
```

```python
#Converting the group by object into a dataframe
group_year_demographics_df = pd.DataFrame(group_year_demographics)
group_year_demographics_df
```

```python
#Final demographics dataframe
final_demographics = group_year_demographics_df.copy()
final_demographics
```

```python
demographicscsv = final_demographics.to_csv('demographics.csv')
```

```python
#Reading the third dataset using pandas
time_series = pd.read_csv("Resources/time_series.csv")
time_series.head()
```

```python
#Retrieving the column names for reference
time_series.columns
```

```python
#Renaming the column names in the standard format that is required in the database
labels2= ['year', 'host_country', 'origin','population_type', 'value']
time_series.columns = labels2
#Replacing the special characters
time_series=time_series.replace(['*'], [0])
time_series
```

```python
#Dropping unwanted column
time_series.drop(['population_type'],axis =1,inplace = True)
time_series
```

```python
#Verifying the datatypes
time_series.dtypes
```

```python
#Converting the data type to numeric for the Value column
time_series['value'] =pd.to_numeric(time_series['value'])
time_series.dtypes
```

```python
#Grouping by year, host country and origin to arrive at the total value
year_group_timeseries = time_series.groupby(['year', 'host_country', 'origin']).sum().reset_index()
year_group_timeseries
```

```python
#Sorting the dataframe to arrive at the total count of asylum seekers over the years base on value(total count)
sorted_timeseries = time_series.sort_values("value", ascending =False).reset_index(drop=True)
sorted_timeseries.head()
```

```python
#Top 25 countries showing the highest number of asylum seekers
final_timeseries = sorted_timeseries[:25]
final_timeseries
```

```python
timeseriescsv = final_timeseries.to_csv('timeseries.csv')
```
